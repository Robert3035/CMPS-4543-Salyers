# 📈 05 - Plotting Basics

Visualization is a core part of data science.  
In this notebook you will learn:
- Setting up Matplotlib for inline plotting
- Creating simple line, scatter, and bar plots
- Customizing titles, labels, and legends
- Why visuals matter in data storytelling


## 1. Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## 2. Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Line Plot Example")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.legend()
plt.show()
```

✅ **Your Turn**: Plot `cos(x)` on the same graph with a different color and label.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y, label="sin(x)")
plt.plot(x, y2, label="cos(x)")

plt.title("Sin and Cos Example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## 3. Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300 * np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot Example")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="color scale")
plt.show()
```

✅ **Your Turn**: Generate two sets of random points and plot them in different colors on the same scatter plot.

```python
x = np.random.rand(50)
y = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.rand(50)
sizes = 300 * np.random.rand(50)
sizes2 = 300 * np.random.rand(50)

plt.scatter(x, y, s=sizes, color='blue', alpha=0.6)
plt.scatter(x2, y2, s=sizes2, color='red', alpha=0.6)
plt.title("Scatter Plot Again")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 4. Bar Plot

```python
categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

✅ **Your Turn**: Create your own bar chart with at least 4 categories (e.g., favorite foods, courses, or hobbies).

```python
categories = ["Baseballs", "Basketballs", "Third category", "Footballs"]
values = [3, 7, 30, 0]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

## 5. Customizing Plots

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="cos(x)", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

✅ **Your Turn**: Customize your own plot with labels, legend, and grid lines.

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.tan(x)
y2 = np.tan(x) * 2

plt.plot(x, y1, label="tan(x)", linestyle="--", color="blue")
plt.plot(x, y2, label="tan(x)*2", linestyle=":", color="red")
plt.title("Customized Line Plot")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()
plt.grid(True)
plt.show()
```

---
### Summary
- `%matplotlib inline` makes plots display inside the notebook.
- Line, scatter, and bar plots are the most common basics.
- Titles, labels, legends, and grids improve readability.
- Visuals tell stories about data more effectively than raw numbers.
